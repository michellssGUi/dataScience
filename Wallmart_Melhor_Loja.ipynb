### 3.1 - Qual o preço médio de vendas semanal por loja?

```python
df.describe(include='all') 
```

```python
total_lojas = 45
media_weekly = []
for i in range(total_lojas):
  venda = []
  for indice, linha in df.iterrows():
    if linha['store'] == i + 1:
      venda.append(linha['weekly_sales'])
  venda_soma = sum(venda)
  media_loja = venda_soma / 143
  media_weekly.append(media_loja)
  print(f'Média das vendas semanais loja_{i + 1}: {media_loja:.2f}') 
```

```python
media_weekly = pd.Series(media_weekly)
media_weekly = enumerate(media_weekly, start=1)
media_weekly = pd.DataFrame(media_weekly)
```

```python
lojas_mais_vende = pd.DataFrame.sort_values(media_weekly, by=1, ascending=False).head()
lojas_mais_vende
```
