### 3.1 - Qual o preço médio de vendas semanal por loja?

```python
total_lojas = 45
total_semanas = 143
media_weekly = []
dicionario_media_weekly = {}
for i in range(total_lojas):
  venda = []
  for indice, linha in df.iterrows():
    if linha['store'] == i + 1:
      venda.append(linha['weekly_sales'])
  venda_soma = sum(venda) # somando o total de vendas realizadas por loja
  media_loja = venda_soma / total_semanas # pegando a média de vendas semanal por loja
  media_weekly.append(media_loja) # esse valor está sendo colhido para realizar tentativa de aderi-lo a tabela posteriormente
  dicionario_media_weekly[i + 1] = dicionario_media_weekly.setdefault(linha['store'] == i + 1, 0) + venda_soma #Colhendo o total de vendas por loja
  #loja_media_vendas = {i + 1 : media_loja}
  print(f'Média das vendas semanais loja_{i + 1}: {media_loja:.2f}') 
```

```python
valor_medio_vendas = media_weekly
pd.Series(valor_medio_vendas)
dd = enumerate(valor_medio_vendas, start=1)
dd = pd.DataFrame(dd)
```

```python
#Loja com mair média de vendas
lojas = pd.DataFrame.sort_values(dd, by=[1], ascending=False)
lojas.head()
```
